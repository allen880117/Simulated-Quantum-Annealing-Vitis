```python
import numpy as np
import math
import matplotlib.pyplot as plt
import statistics as st
import os, time
```

```python
rec_basic = []
rec_basic_detail = []
basic_log_path = '../../../sqa/solution1/csim/build/energy.naive.txt'
print(time.ctime(os.path.getmtime(basic_log_path)))
with open(basic_log_path, 'r') as file:
    tmp = []
    for line in file:
        if ('T' in line):
            tmp.append(float(line.split()[1]))
        if ('Energy:' in line):
            rec_basic.append(float(line.split()[1]))
            # rec_basic.append(np.min(tmp))
            rec_basic_detail.append(tmp)
            tmp = []

rec_u50 = []
rec_u50_detail = []
u50_log_path = '../../../sqa/solution1/csim/build/energy.u50.txt'
print(time.ctime(os.path.getmtime(u50_log_path)))
with open(u50_log_path, 'r') as file:
    tmp = []
    for line in file:
        if ('T' in line):
            tmp.append(float(line.split()[1]))
        if ('Energy:' in line):
            rec_u50.append(float(line.split()[1]))
            # rec_u50.append(np.min(tmp))
            rec_u50_detail.append(tmp)
            tmp = []

print(np.min((rec_basic)))
print(np.min((rec_u50)))
```

```python
plt.figure(figsize=(30,10))
plt.plot(np.array(rec_basic[:])+1, 'o-')
plt.plot(np.array(rec_u50[:]), 'o-')
print(np.sum(np.array(rec_basic[:])-np.array(rec_u50[:])) == 0.0)
```

```python
len_record = np.min([len(rec_basic_detail), len(rec_u50_detail)])
len_trot = np.min([len(rec_basic_detail[0]), len(rec_u50_detail[0])])
all_same = True
err_idx = []
for i in range(len_record):
    for j in range(len_trot):
        if rec_basic_detail[i][j] != rec_u50_detail[i][j]:
            all_same = False
            err_idx.append((i, j))
print(all_same)        
```
